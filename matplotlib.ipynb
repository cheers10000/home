```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(t1,f(t1),"bo",t2,f(t2),"k")
plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),"r--")
plt.show()
```

```python
import matplotlib.mlab as mlab
from scipy.stats import norm
fig=plt.figure(figsize=(10,8))
ax1=fig.add_subplot(111)
n,bins,patches=ax1.hist(df[""],bins=100,color="m")
ax1.set_ylabel("",fontsize=15)
ax1.set_ylabel("",fontsize=15)
ax1.set_title("",fontsize=20)
y=norm.pdf(bins,df[""].mean(),df[""].std())
ax2=ax1.twinx()
ax2.plot(bins,y,"b--")
ax2.set_ylabel("",fontsize=15)
plt.show()
```

```python
x=df[""][::100]
y=df[""][::100]
plt.figure(figsize=(10,6))
plt.scatter(x,y,color="c",marker="d")
plt.legend()
plt.title("",fontsize=20)
plt.xlabel("",fontsize=18)
plt.ylabel("",fontsize=18)
plt.show()
```

```python
data=df[df.==""][""]
plt.figure(figsize=(10,6))
plt.boxplot(data,whis=2,flierprops={"marker":"o","markerfacecolor":"r","color":"k"},
           patch_artist=True,boxprops={"color":"k","facecolor":"#9999ff"})
plt.title("",fontsize=20)
plt.show()
```

```python
data1=df[df.==""][""]
data2=df[df.==""][""]
data3=df[df.==""][""]
data4=df[df.==""][""]
data5=df[df.==""][""]
plt.figure(figsize=(12,8))
plt.boxplot([data1,data2,data3,data4,data5],labels=["","","","",""],
           whis=2,flierprops={"marker":"o","markerfacecolor":"r","color":"k"},
           patch_artist=True,boxprops={"color":"k","facecolor":"#9999ff"},
           vert=False)
ax=plt.gca()
ax.patch.set_facecolor("gray")
ax.patch.set_alpha(0.3)
plt.title("",fontsize=20)
plt.show()

```
